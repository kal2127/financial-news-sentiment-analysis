### Basic Bar, Line, and Scatter Plots

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr","May","June","July","Aug","Sep","Oct","Nov","Dec"],
    "Sales": [1200, 1800, 900, 1500, 1000, 1600, 1200, 800, 1500, 1700, 1400, 1000]
}

df = pd.DataFrame(data)
```

```python

# Line chart
plt.figure()
plt.plot(df["Month"], df["Sales"])
plt.title("Monthly Sales Trend")
plt.show()

```

```python
# Bar chart
plt.figure()
plt.bar(df["Month"], df["Sales"])
plt.title("Monthly Sales Comparison")
plt.show()
```

```python

# Scatter plot
plt.figure()
plt.scatter(df["Month"], df["Sales"])
plt.title("Scatter: Sales")
plt.show()
```

```python
print(df.columns)
```

```python
# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(df["Sales"], bins=5)
plt.title("Histogram of Monthly Sales")
plt.xlabel("Sales")
plt.ylabel("Frequency")

plt.show()
```

### Histogram

How data distributions look, How variability is shown visually

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(loc=50, scale=10, size=500)

plt.hist(data, bins=20)
plt.title("Distribution of Values")
plt.show()
```

### Correlation (Scatter + Trendline)
How two variables relate, How to visually detect correlation

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.randint(1, 100, 50)
y = x * 2 + np.random.randint(1, 20, 50)  # correlated data

plt.scatter(x, y)
plt.title("Correlation Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

### Misleading Visualization
How charts can lie, Why axis manipulation matters

```python
import matplotlib.pyplot as plt

x = ["A", "B"]
y = [20, 25]

# Correct chart
plt.figure()
plt.bar(x, y)
plt.title("Correct Axis")
plt.ylim(0, 30)
plt.show()

# Misleading chart
plt.figure()
plt.bar(x, y)
plt.title("Misleading Axis (Zoomed In)")
plt.ylim(19, 26)
plt.show()
```

### High Data-Ink vs Low Data-Ink
- Tuftes Data-Ink Ratio
- Why removing unnecessary elements improves clarity

#### High Data-Ink(good)

```python
import matplotlib.pyplot as plt

x = ["A", "B", "C", "D"]
y = [5, 8, 2, 7]

plt.bar(x, y)
plt.title("High Data-Ink")
plt.show()
```

#### Low Data-Ink (Bad)

```python
import matplotlib.pyplot as plt

plt.figure()
plt.bar(x, y, color="skyblue")
plt.grid(True)
plt.title("Bad Chart with Extra Noise")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

#### Using Subplots to Compare Data
- How dashboards compare multiple variables
- Layout and storytelling

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 11)
y1 = np.random.randint(5, 15, 10)
y2 = np.random.randint(10, 30, 10)

fig, axs = plt.subplots(1, 2, figsize=(10,4))

axs[0].plot(x, y1)
axs[0].set_title("Product A Sales")

axs[1].bar(x, y2)
axs[1].set_title("Product B Sales")

plt.show()
```
