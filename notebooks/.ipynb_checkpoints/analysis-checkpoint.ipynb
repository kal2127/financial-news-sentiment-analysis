```python
import pandas as pd
import numpy as np

# Load your dataset
df = pd.read_csv("../src/raw_analyst_ratings.csv")

# Show the first 5 rows
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python

# Define the source timezone (UTC-4 is typically America/New_York / EST)
TIMEZONE = 'America/New_York'

# 1. Clean the string column: Remove the timezone offset (e.g., '-04:00') if it exists,
# to force a consistent, naive date string format (YYYY-MM-DD HH:MM:SS).
# This regex removes '+/-HH:MM' at the end of the string.
df['date_clean'] = df['date'].str.replace(r'[+-]\d{2}:\d{2}$', '', regex=True).str.strip()

# 2. Convert to Naive Datetime
# Use a specific format string for explicit parsing.
df['datetime_naive'] = pd.to_datetime(
    df['date_clean'], 
    format='%Y-%m-%d %H:%M:%S', 
    errors='coerce' # Set unparsable dates to NaT instead of crashing
)

# 3. Localize to Source Timezone (America/New_York)
# We assume all news originated in this timezone (UTC-4).
df['datetime_local'] = df['datetime_naive'].dt.tz_localize(
    TIMEZONE, 
    nonexistent='shift_forward', 
    ambiguous='NaT'
)

# 4. Standardize Timezone to UTC (Final, clean column for analysis)
# This is the definitive column for all time-series analysis.
df['datetime_utc'] = df['datetime_local'].dt.tz_convert('UTC')

# 5. Extract components for EDA
df['publication_date'] = df['datetime_utc'].dt.date
df['day_of_week'] = df['datetime_utc'].dt.day_name()
df['hour_of_day'] = df['datetime_utc'].dt.hour

# Show the cleaned result
print(df[['date', 'datetime_utc', 'day_of_week', 'hour_of_day']].head())
```

```python
# Headline length (characters)
df['headline_length'] = df['headline'].astype(str).apply(len)

# Word count
df['headline_word_count'] = df['headline'].astype(str).apply(lambda x: len(x.split()))

# Extract date only from the CORRECT, cleaned column
# Note: I am assuming the previous fix was run, which created 'datetime_utc'
df['date_only'] = df['datetime_utc'].dt.date

df[['headline_length', 'headline_word_count', 'date_only']].head()
```

```python
# Summary of headline length
print("Headline Length Statistics:")
print(df['headline_length'].describe())

print("\nWord Count Statistics:")
print(df['headline_word_count'].describe())
```

```python
import pandas as pd
import numpy as np

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from textblob import TextBlob

from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
```

```python
## Cell 1.5: VADER Data Download (Run once)

import nltk
nltk.download('vader_lexicon')
```

```python
## Cell 3: VADER and TextBlob Sentiment (Optimized)

# You must have run Cell 1 (Imports) and Cell 2 (Data Loading) first.
# You may need to run: pip install tqdm

import tqdm
from textblob import TextBlob
from nltk.sentiment.vader import SentimentIntensityAnalyzer
tqdm.tqdm.pandas()

# --- VADER Functions ---
analyzer = SentimentIntensityAnalyzer()
def get_vader_score(text):
    """Calculates VADER compound score."""
    try:
        return analyzer.polarity_scores(str(text))['compound']
    except:
        return 0.0

# --- TextBlob Functions ---
def get_tb_polarity(text):
    """Calculates TextBlob polarity score."""
    try:
        blob = TextBlob(str(text))
        return blob.sentiment.polarity
    except:
        return 0.0

# --- Apply Functions with Progress Bar ---
print("Calculating VADER scores...")
df['vader_score'] = df['headline'].astype(str).progress_apply(get_vader_score)

print("Calculating TextBlob scores...")
df['tb_polarity'] = df['headline'].astype(str).progress_apply(get_tb_polarity)

print("VADER and TextBlob scores created successfully.")
print(df[['vader_score', 'tb_polarity']].head())
```

```python
tokenizer = AutoTokenizer.from_pretrained("ProsusAI/finbert")
model = AutoModelForSequenceClassification.from_pretrained("ProsusAI/finbert")
```

```python
## Cell 4: FinBERT Sentiment (Self-Contained & Optimized)

# --- 0. Progress Bar Setup (REQUIRED) ---
import tqdm
import torch # Ensure torch is imported for FinBERT
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# Activate the progress_apply method
tqdm.tqdm.pandas() 

# --- 1. FINBERT MODEL SETUP ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
tokenizer = AutoTokenizer.from_pretrained("ProsusAI/finbert")
model = AutoModelForSequenceClassification.from_pretrained("ProsusAI/finbert").to(device)


# --- 2. FINBERT SCORING FUNCTION ---
def finbert_sentiment(text):
    """Calculates FinBERT score (Positive - Negative probability)."""
    try:
        # Tokenize and move to device (CPU/GPU)
        inputs = tokenizer(str(text), return_tensors="pt", padding=True, truncation=True, max_length=128).to(device)
        
        # Run model inference
        with torch.no_grad():
            outputs = model(**inputs)
            
        scores = torch.softmax(outputs.logits, dim=1).cpu().numpy()[0]
        
        # Score = Probability(Positive) - Probability(Negative)
        return scores[2] - scores[0]
    except Exception as e:
        return 0.0

# --- 3. APPLY FUNCTION WITH PROGRESS BAR ---

print(f"Starting FinBERT calculation on {device}...")
# The progress bar will now show up here:
df['finbert_score'] = df['headline'].astype(str).progress_apply(finbert_sentiment)

print("FinBERT calculation complete.")
print(df[['headline', 'finbert_score']].head())
```

```python
# --- TextBlob Sentiment ---
from textblob import TextBlob

def textblob_sentiment(text):
    try:
        blob = TextBlob(str(text))
        return blob.sentiment.polarity
    except:
        return 0

df['tb_polarity'] = df['headline'].apply(textblob_sentiment)


# --- FinBERT Placeholder (until you install FinBERT) ---
# Replace this with real FinBERT predictions later
df['finbert_score'] = 0.0


# --- Final Weighted Sentiment Score ---
df['sentiment_final'] = (
    df['vader_score']*0.3 +     # VADER (already created earlier)
    df['tb_polarity']*0.2 +         # TextBlob
    df['finbert_score']*0.5         # FinBERT (0 for now)
)

df[['headline', 'sentiment_score', 'tb_polarity', 'sentiment_final']].head()
```

```python
plt.hist(df['sentiment_final'], bins=30)
plt.title("Sentiment Score Distribution")
plt.xlabel("Sentiment Score")
plt.ylabel("Frequency")
plt.show()

```
